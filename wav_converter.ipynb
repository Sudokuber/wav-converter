```python
import numpy as np
from scipy.io.wavfile import read, write
import warnings
warnings.filterwarnings("ignore")
from pathlib import Path
```

```python
def wav_converter(file_name, out_file_name, depth, new_depth, sign, new_sample_rate):
    
    sample_rate, data = read(file_name)
    
    # -----------Sample Depth---------------------------------------
    # bits shift
    new_data = data >> (int(depth) - int(new_depth))
    
    # -----------Convert from int to uint---------------------------
    if sign == False:
        new_data[new_data < 0] = new_data[new_data < 0] + (2 ** int(new_depth))

    # -----------Sample Rate----------------------------------------
    # No.second = total number of samples / samples per second
    seconds = data.size / sample_rate
    # calculate the new total number of samples 
    new_data_size = round(seconds * new_sample_rate)
    # calculate the new distance between two samples
    dist = np.round(np.linspace(0, data.size-1, new_data_size))
    dist = dist.astype(int)
    # resample 
    new_data = new_data[dist]
    
    # ----------Write to file--------------------------------------
    dir = Path.cwd().joinpath('/'+ out_file_name)
    with open(dir, 'w') as f:
        f.write('data[{}] = {{'.format(len(new_data)))
        count = 0
        for i in new_data:
            if count % 20 != 0:
                f.write('%s' %i + ', ')
            else:
                f.write('\n' + '%s' %i + ', ' )
            count += 1
        f.write('}')
    


        
```

```python
Go = True

while Go:
    print("To quit press 'q'")
    
#-> ask for input file name
    file_name = input('Name of the .wav file that needs to be converted: ')
    if file_name in ['q','Q']:
        print('Exit')
        break
    else:
        if '.wav' not in file_name:
            file_name = file_name + '.wav'
        
#-> name output file 
    out_file_name = input('Name the output text file: ')
    if out_file_name in ['q', 'Q']:
        print('Exit')
        break
    else:
        if '.txt' not in out_file_name:
            out_file_name = out_file_name + '.txt'

#-> ask current depth
    depth = input('Current sample depth (eg. 16bits is 16): ')
    if depth in ['q', 'Q']:
        print('Exit')
        break
    else:
        if len(depth) > 2:
            print('value is not valid, try again.')
            depth = input('Current sample depth (eg. 16bits is 16): ')
        
#-> ask new depth & sign/unsigned
    new_depth = input('New sample depth (eg. 8bits is 8): ')
    if new_depth in ['q', 'Q']:
        print('Exit')
        break
    else:
        if len(new_depth) > 2:
            print('Value is not valid, try again.')
            depth = input('Current sample depth (eg. 16bits is 16): ')
        
    sign = input('new depth is signed or unsigned (s / u): ')
    if new_depth in ['q', 'Q']:
        print('Exit')
        break
    else:
        if sign in ['s', 'Sign', 'sign', 'S']:
            sign = True
        else: 
            sign = False
        
#-> ask new sample rate
    new_sample_rate = input('New sample rate (eg. 16kHz is 16000): ')
    if new_sample_rate in ['q', 'Q']:
        print('Exit')
        break
    else:
        if len(new_sample_rate) < 4:
            print('Value is not valid, try again.')
            depth = input('Current sample depth (eg. 16bits is 16): ')
        new_sample_rate = int(new_sample_rate)
    
    wav_converter(file_name, out_file_name, depth, new_depth, sign, new_sample_rate)
    print('File is converted and saved.')
    print('Exit')
    
    Go = False

    
```
